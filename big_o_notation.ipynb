```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
n=[n for n in range(1,100,10)]
y1=[(0 * n) + 1 for n in n]
y2=[np.log(n) for n in n]
y3=[n for n in n]
y4=[(n * np.log(n)) for n in n]
y5=[n ** 2 for n in n]
y6=[2 ** n for n in n]
y7=[np.math.factorial(n) for n in n]
y8=[n ** n for n in n]

plt.plot(n,y1,c='violet',label='$y=O(1)$')
plt.plot(n,y2,c='indigo',label='$O(\log n)$')
plt.plot(n,y3,c='blue',label='$O(n)$')
plt.plot(n,y4,c='green',label='$O(n \log n)$')
plt.plot(n,y5,c='yellow',label='$O(n^2)$')
plt.plot(n,y6,c='orange',label='$O(2^n)$')
plt.plot(n,y7,c='red',label='$O!$')
plt.plot(n,y8,c='crimson',label='$O(n^n)$')

plt.ylim(0,200)
plt.xlim(-1,90)

plt.title('Big O Notation')

plt.xlabel('Elements')
plt.ylabel('Operations')

plt.legend()

plt.show()
```
